```python
# Using Hard-Coded Variables
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(time):
  # Hard-Coded coefficients for the quadratic equation
  a=0.1
  b=-1
  c=30
  # Quadratic Equation representing temperature change
  temperature = a* (time **2) + b*time + c
  return temperature

def main():
  # Time values from 0 to 10 (for example)
  time_values = np.linspace (0,10,50)
  # Calculate temperatures using hard-coded coefficients
  temperature_hardcoded = quadratic_model(time_values)
  # Plot the results
  plt.plot(time_values, temperature_hardcoded, label=f'Hard-Coded Coefficients: a={0.1}, b={-1}, c={30}')
  plt.xlabel('Time')
  plt.ylabel('Temperature')
  plt.legend()
  plt.title('Weather Modelling with Quadratic Equation (Hard-Coded Coefficients)' )
  plt.show()

if __name__=='__main__':
  main()




```

```python
# Using Single set of input values
import numpy as np
import matplotlib.pyplot as plt

# Define the quadratic model function
def quadratic_model(time, a, b, c):
  temperature = a * time**2 + b * time + c
  return temperature

# Define the main function
def main():
  # Define the input coefficients as a list
  coefficients = [0.1, -1, 30]
  # Time values from 0 to 10 (for example)
  time_values = np.linspace(0, 10, 50)
  # Calculate temperatures using the coefficients from the list
  temperature = quadratic_model(time_values, *coefficients)
  # Plot the results
  plt.plot(time_values, temperature, label=f'Single set of values: a={coefficients[0]}, b={coefficients[1]}, c={coefficients[2]}')
  plt.xlabel('Time')
  plt.ylabel('Temperature')
  plt.legend()
  plt.title('Weather Modelling with Quadratic Equation (Single set of values)')
  plt.show()

if __name__ == '__main__':
  main()
```

```python
# Using Multiple sets of input values
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(time, a, b, c):
    temperature = a * (time ** 2) + b * time + c
    return temperature

def main():
    # Define multiple sets of coefficients
    coefficients = [
        [0.1, -1, 30],
        [0.5, -2, 25],
        [1, -3, 20],
    ]
    # Define the time values
    time_values = np.linspace(0, 10, 50)
    # Plot the results for each set of coefficients
    for i, coeffs in enumerate(coefficients):
        a, b, c = coeffs
        temperature = quadratic_model(time_values, a, b, c)
        plt.plot(time_values, temperature, label=f'Set {i + 1}: a={a}, b={b}, c={c}')
    # Add labels and title
    plt.xlabel("Time")
    plt.ylabel("Temperature")
    plt.legend()
    plt.title("Weather Modeling with Quadratic Equation (Multiple sets of values)")
    plt.show()

if __name__ == "__main__":
    main()
```

```python
# Using Multiple set of file input values
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(time, a, b, c):
  temperature = a * (time ** 2) + b * time + c
  return temperature

def main():
  # Open the input file
  with open('/content/coefficients1.txt', 'r') as f:
    coefficients = []
    for line in f:
      # Extract coefficients from each line
      a, b, c = map(float, line.split())
      coefficients.append([a, b, c])
  # Define the time values
  time_values = np.linspace(0, 10, 50)
  # Plot the results for each set of coefficients
  for i, coeffs in enumerate(coefficients):
    a, b, c = coeffs
    temperature = quadratic_model(time_values, a, b, c)
    plt.plot(time_values, temperature, label=f'Set {i + 1}: a={a}, b={b}, c={c}')
  # Add labels and title
  plt.xlabel("Time")
  plt.ylabel("Temperature")
  plt.legend()
  plt.title("Weather Modeling with Quadratic Equation (Multiple Sets of Inputs from File)")
  plt.show()

if __name__ == "__main__":
  main()
```

```python
# Using a combination of Hard-Coded and Keyboard Input values
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(time, a, b, c):
  temperature = a * (time ** 2) + b * time + c
  return temperature

def main():
  # Time values from 0 to 10 (for example)
  time_values = np.linspace(0, 10, 50)
  # Hard-coded coefficients for the first curve
  a1 = 15
  b1 = 45
  c1 = 60
  # Keyboard input coefficients for the second curve
  a2 = float(input("Enter the value of 'a' for the second curve: "))
  b2 = float(input("Enter the value of 'b' for the second curve: "))
  c2 = float(input("Enter the value of 'c' for the second curve: "))
  # Calculate temperatures for both curves
  temperature_hardcoded = quadratic_model(time_values, a1, b1, c1)
  temperature_keyboard = quadratic_model(time_values, a2, b2, c2)
  # Plot the results
  plt.plot(time_values, temperature_hardcoded, label=f'Hard-Coded Coefficients: a={a1}, b={b1}, c={c1}')
  plt.plot(time_values, temperature_keyboard, label=f'Keyboard Input Coefficients: a={a2}, b={b2}, c={c2}')
  plt.xlabel('Time')
  plt.ylabel('Temperature')
  plt.legend()
  plt.title('Weather Modeling with Quadratic Equation (combination of Two Curves)')
  plt.show()

if __name__ == '__main__':
  main()
```
